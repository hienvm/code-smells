```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
import pickle

file_path = './results.pkl'
with open(file_path, "rb") as fh:
  results_df = pickle.load(fh)

results_df.columns = ['dataset', 'random_seed', 'train_f1_micro', 'train_f1_macro', 'train_accuracy', 'classification_report']
results_df.head()
```

```python
# parse dataset name
results_df.dataset = [x.split('/')[-1].split('.')[0] for x in results_df.dataset]
results_df.classification_report = [x.split() for x in results_df.classification_report]
results_df.head()
```

```python
# parse classification report
results_df['precision_0'] = [float(x[5]) for x in results_df.classification_report]
results_df['recall_0'] = [float(x[6]) for x in results_df.classification_report]
results_df['f1_score_0'] = [float(x[7]) for x in results_df.classification_report]

results_df['precision_1'] = [float(x[10]) for x in results_df.classification_report]
results_df['recall_1'] = [float(x[11]) for x in results_df.classification_report]
results_df['f1_score_1'] = [float(x[12]) for x in results_df.classification_report]

results_df['accuracy'] = [float(x[15]) for x in results_df.classification_report]

results_df['precision_macro_avg'] = [float(x[19]) for x in results_df.classification_report]
results_df['recall_macro_avg'] = [float(x[20]) for x in results_df.classification_report]
results_df['f1_score_macro_avg'] = [float(x[21]) for x in results_df.classification_report]

results_df['precision_weighted_avg'] = [float(x[25]) for x in results_df.classification_report]
results_df['recall_weighted_avg'] = [float(x[26]) for x in results_df.classification_report]
results_df['f1_score_weighted_avg'] = [float(x[27]) for x in results_df.classification_report]

results_df.drop(columns=['classification_report'], inplace=True)

results_df.head()
```
