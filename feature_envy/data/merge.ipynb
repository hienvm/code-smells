```python
import pandas as pd
```

```python
df = pd.read_pickle('../data/embedded_datasets/T5_plus_whole.pkl')
df = df.drop(columns=['embeded_sequence_sum','embeded_sequence_avg'])
pd.to_pickle(df, '../data/embedded_datasets/T5_plus_whole.pkl')
```

```python
aggregations = ['whole', 'avg', 'sum']
for agg in aggregations:
    df_t5p = pd.read_pickle(f'../data/embedded_datasets/T5_plus_{agg}.pkl')
    df_t5p = df_t5p.drop(columns=['severity', 'label'])
    df_metrics = pd.read_pickle('../data/embedded_datasets/metrics_dataset.pkl')
    df = pd.merge(df_t5p, df_metrics, on='sample_id')
    pd.to_pickle(df, f'./embedded_datasets/{agg}_merged.pkl')
```
